### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
def calcFaturamento(loja):
  return round(sum(loja['Preço']), 2)
```

```python
# Faturamento - Loja 1

faturamento_1 = calcFaturamento(loja)
print(f"R$ {faturamento_1:,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.'))
```

```python
# Faturamento - Loja 2

faturamento_2 = calcFaturamento(loja2)
print(f"R$ {faturamento_2:,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.'))
```

```python
# Faturamento - Loja 3

faturamento_3 = calcFaturamento(loja3)
print(f"R$ {faturamento_3:,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.'))
```

```python
# Faturamento - Loja 4

faturamento_4 = calcFaturamento(loja4)
print(f"R$ {faturamento_4:,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.'))
```

```python
df_faturamento = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Faturamento': [faturamento_1, faturamento_2, faturamento_3, faturamento_4]
})

df_faturamento.sort_values('Faturamento', ascending=False)

display(df_faturamento)
```

# 2. Vendas por Categoria

```python
def vendasCategoria(loja):
  return loja.groupby('Categoria do Produto').sum('Preço').sort_values('Preço', ascending=False)

```

```python
# Vendas por categoria = Loja 1

vendasCategoria(loja)
```

```python
# Vendas por categoria = Loja 2

vendasCategoria(loja2)
```

```python
# Vendas por categoria = Loja 3

vendasCategoria(loja3)
```

```python
# Vendas por categoria = Loja 4

vendasCategoria(loja4)
```

# 3. Média de Avaliação das Lojas


```python
mediaLoja1 = loja['Avaliação da compra'].mean().round(2)
mediaLoja2 = loja2['Avaliação da compra'].mean().round(2)
mediaLoja3 = loja3['Avaliação da compra'].mean().round(2)
mediaLoja4 = loja4['Avaliação da compra'].mean().round(2)

mediaAvaliacao = pd.DataFrame({'Loja 1': mediaLoja1, 'Loja 2': mediaLoja2,
                   'Loja 3': mediaLoja3, 'Loja 4': mediaLoja4}, index=['Média'])

mediaAvaliacao
```

# 4. Produtos Mais e Menos Vendidos

```python
lista_lojas = [loja, loja2, loja3, loja4]
nomes_loja = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
relatorio_por_loja = {}

for i, df_loja in enumerate(lista_lojas):
  ranking = vendasCategoria(df_loja)

  mais_vendidos = ranking.head(5)
  menos_vendidos = ranking.tail(5)

  relatorio_por_loja[nomes_loja[i]] = pd.concat([mais_vendidos, menos_vendidos], keys=['Mais vendidos', 'Menos vendidos'])
```

# 5. Frete Médio por Loja

```python
def frete_medio_loja(df_loja):
  return df_loja[df_loja['Frete'] > 0]['Frete'].mean().round(2)

frete_loja1 = frete_medio_loja(loja)
frete_loja2 = frete_medio_loja(loja2)
frete_loja3 = frete_medio_loja(loja3)
frete_loja4 = frete_medio_loja(loja4)

frete_lojas = [frete_loja1, frete_loja2, frete_loja3, frete_loja4]

frete_medio = pd.DataFrame({'Lojas': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
              'Frete médio': [frete_loja1, frete_loja2, frete_loja3, frete_loja4]})

display(frete_medio)

```

#### 1. Faturamento por loja

```python
axes1 = df_faturamento.plot(kind='bar', x='Loja', y='Faturamento', color='skyblue', figsize=(10, 6), legend=False)

# Encontra as barras da lista de containers de índice 0, resgata a altura de
# cada patche (as barras) e atribui um label formatado pra cada
axes1.bar_label(axes1.containers[0], padding=3, fmt='R$ {:,.2f}')

# Eixo y vai de 0 até o valor da maior barra + 15%, adicionando espaçamento
# para o label
plt.ylim(0, df_faturamento['Faturamento'].max() * 1.15)

plt.title('Faturamento por loja')
plt.ylabel('R$ (milhões)')
plt.xticks(rotation=0)



plt.show()
```

#### 2. Distribuição de vendas por categoria

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

axes = axes.flatten()

for i, nome_loja in enumerate(nomes_loja):

  dados = relatorio_por_loja[nome_loja].loc['Mais vendidos']

  values = dados['Preço']
  labels = dados.index

  axes[i].pie(
      values,
      labels=labels,
      autopct='%1.1f%%'
  )

  faturamento = df_faturamento['Faturamento'][i]
  titulo = f"{nome_loja} | R$ {faturamento:,.2f}"
  axes[i].set_title(titulo)

fig.suptitle('Distribuição de vendas por categoria', fontsize=16)
plt.tight_layout(pad=5.0)
```

#### 3. Comparação de Categorias: Lojas Líderes vs. Loja 4

```python
import numpy as np

eletronicos_faturamento = []
eletrodomesticos_faturamento = []

for nome in nomes_loja:
  eletronicos_faturamento.append(relatorio_por_loja[nome].loc['Mais vendidos'].loc['eletronicos']['Preço'])
  eletrodomesticos_faturamento.append(relatorio_por_loja[nome].loc['Mais vendidos'].loc['eletrodomesticos']['Preço'])

grupos = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

x = np.arange(len(grupos))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 7))
plt.title('Comparação de Categorias: Lojas Líderes vs. Loja 4')

rects1 = ax.bar(x - width/2, eletronicos_faturamento, width, label='Eletrônicos', color='Orange')
rects2 = ax.bar(x + width/2, eletrodomesticos_faturamento, width, label='Eletrodomésticos', color='Blue')

for container in ax.containers:
  ax.bar_label(container, fmt='R$ {:,.0f}')

ax.set_xticks(x)
ax.set_xticklabels(grupos)
plt.ylim(0, max(eletronicos_faturamento) * 1.15)

ax.legend()
plt.show()
plt.savefig('comparacao_categorias.png')
```
